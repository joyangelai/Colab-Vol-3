### Checking annual trend using Resample technique with rules

```python
#checking annual mean
data.resample(rule="A").mean()["High"]
```

```python
#ploting the above
data.resample(rule="A").mean()["High"].plot(figsize=(12,4))
```

```python
#plort using the bar plot
data.resample(rule="A").mean()["High"].plot(kind= 'bar', figsize=(12,4));
```

Checking monthly trend using Resample technique with rules

```python
#monthly trend
data.resample(rule="M").mean()["High"]
```

```python
#plot
data.resample(rule="M").mean()["High"].plot(figsize=(12,4))
```

```python
#plot
data.resample(rule="q").mean()["High"].plot(figsize=(12,4))
```

#### Time Series Smoothening



```python
#start with the original plot
data["High"].plot(figsize=(12,4));
```

####1. Simple moving average

Cons: it give priority to old data

```python
#Simple moving average(rolling)
data["high sma 10"]=data["High"].rolling(window=10).mean()

data.head(13)

```

```python
data[["High","high sma 10"]].plot(figsize=(12,4))
```

```python
data["high sma 5"]=data["High"].rolling(window=5).mean()
```

```python
data[["High","high sma 10", "high sma 5"]].plot(figsize=(12,4))
```

#### 2. Commulative moving average (expanding)

```python
#CMA
data["High"].expanding().mean().plot()
```

#### 3. Expontential moving average

```python
#EWA
data["High ewa 10"]=data["High"].ewm(span=10).mean()
```

```python
data[["High","High ewa 10"]].plot(figsize=(12,4))
```

### 4. Expotential moving weighted moving average

```python
#emwa
data["High ewma"]=data["High"].ewm(alpha=0.5).mean()
```

```python
data[["High","High ewma"]].plot(figsize=(12,4))
```
